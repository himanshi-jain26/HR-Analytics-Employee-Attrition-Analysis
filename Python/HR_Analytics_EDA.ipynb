# Attrition Analysis

## Attrition Distribution

```python
plt.figure(figsize=(6,4))

df['Attrition'].value_counts().plot(
    kind='bar'
)

plt.title(
    'Employee Attrition Distribution'
)

plt.xlabel(
    'Attrition'
)

plt.ylabel(
    'Employee Count'
)

plt.show()
```

The organization experienced an attrition rate of approximately 16%, indicating that most employees remain with the company. However, employee turnover still represents a significant business concern.

## Attrition by Department

```python
department_attrition = pd.crosstab(
    df['Department'],
    df['Attrition']
)

department_attrition
```

## Plot

```python
department_attrition.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title(
    'Attrition by Department'
)

plt.xlabel(
    'Department'
)

plt.ylabel(
    'Employees'
)

plt.show()
```

Research & Development recorded the highest employee attrition, followed by Sales. These departments should be prioritized for retention and employee engagement initiatives.

## Attrition by Gender

```python
gender_attrition = pd.crosstab(
    df['Gender'],
    df['Attrition']
)

gender_attrition.plot(
    kind='bar',
    figsize=(6,4)
)

plt.title(
    'Attrition by Gender'
)

plt.show()
```

Male employees experienced a higher number of exits compared to female employees, suggesting demographic differences in retention patterns.

## Attrition by Overtime

```python
overtime_attrition = pd.crosstab(
    df['OverTime'],
    df['Attrition']
)

overtime_attrition.plot(
    kind='bar',
    figsize=(6,4)
)

plt.title(
    'Attrition by Overtime'
)

plt.show()
```

Overtime appears to be a major driver of employee attrition. Employees working overtime show substantially higher turnover compared to those with standard working hours.

# Employee Analysis

## Attrition by Job Role

```python
jobrole_attrition = pd.crosstab(
    df['JobRole'],
    df['Attrition']
)

jobrole_attrition.plot(
    kind='bar',
    figsize=(12,5)
)

plt.title(
    'Attrition by Job Role'
)

plt.show()
```

Laboratory Technicians and Sales Executives experienced the highest employee turnover, while managerial and senior research roles demonstrated stronger retention.

## Average Income by Department

```python
income_dept = df.groupby(
    'Department'
)['MonthlyIncome'].mean()

income_dept
```

## Plot

```python
income_dept.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title(
    'Average Monthly Income by Department'
)

plt.ylabel(
    'Income'
)

plt.show()
```

Employees in the Sales department earned the highest average monthly income, while Research & Development employees received comparatively lower compensation.

## Age Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(
    df['Age'],
    bins=10
)

plt.title(
    'Employee Age Distribution'
)

plt.xlabel(
    'Age'
)

plt.ylabel(
    'Employees'
)

plt.show()
```

The workforce is predominantly composed of mid-career professionals aged between 30 and 40 years.

## Job satisfaction

```python
df[
    'JobSatisfaction'
].value_counts().sort_index().plot(
    kind='bar'
)

plt.title(
    'Job Satisfaction Distribution'
)

plt.show()
```

Overall job satisfaction appears relatively positive, with the majority of employees reporting moderate to high satisfaction levels.

# Advanced Employee Retention Analysis

## Attrition Rate by Department

```python
dept_attrition = (
    pd.crosstab(
        df['Department'],
        df['Attrition'],
        normalize='index'
    ) * 100
)

dept_attrition
```

## Plot

```python
dept_attrition['Yes'].sort_values(
    ascending=False
).plot(
    kind='bar',
    figsize=(8,5)
)

plt.title(
    'Attrition Rate by Department'
)

plt.ylabel(
    'Attrition Rate (%)'
)

plt.show()
```

The Sales department experienced the highest attrition rate (20.63%), indicating potential challenges related to workload, performance pressure, or employee engagement. Research & Development showed the lowest attrition rate (13.84%), suggesting better employee retention.

## Attrition Rate by Job Role

```python
job_attrition_rate = (
    pd.crosstab(
        df['JobRole'],
        df['Attrition'],
        normalize='index'
    ) * 100
)

job_attrition_rate['Yes'].sort_values(
    ascending=False
).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title(
    'Attrition Rate by Job Role'
)

plt.ylabel(
    'Attrition Rate (%)'
)

plt.show()
```

Sales Representatives experienced the highest attrition rate, with nearly 40% of employees leaving the organization. Senior roles such as Research Director and Manager demonstrated significantly stronger retention.

## Attrition by Marital Status

```python
pd.crosstab(
    df['MaritalStatus'],
    df['Attrition']
).plot(
    kind='bar',
    figsize=(8,5)
)

plt.title(
    'Attrition by Marital Status'
)

plt.show()
```

Single employees appear more likely to leave the organization than married employees. This may reflect differences in career mobility, job expectations, or personal commitments.

## Attrition by Education Field

```python
pd.crosstab(
    df['EducationField'],
    df['Attrition']
).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title(
    'Attrition by Education Field'
)

plt.show()
```

Employees from Life Sciences and Medical education backgrounds account for the largest share of employee exits. Since these groups also represent a large portion of the workforce, targeted retention programs may help reduce turnover.

## Work-Life Balance Distribution

```python
df[
    'WorkLifeBalance'
].value_counts().sort_index().plot(
    kind='bar'
)

plt.title(
    'Work Life Balance Distribution'
)

plt.show()
```

Most employees reported a moderate work-life balance, suggesting that work-life policies are generally effective. However, a smaller group reporting poor balance may represent a higher attrition risk.
